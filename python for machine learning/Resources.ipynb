### Data wrangling

```python
data_1 = data.head(5)
```

```python
data_1
```

```python
data_2 = data.tail(10)
```

```python
data_2
```

```python
pd.merge(data_1, data_2, how='outer')
```

> Note:
    The merge method can be done using three different methods:
    
        1. outer: The outer method acts just like the SQL full outer join which 
        merges values of both datasets and presents them as one
        
        2. inner: The inner method acts just like the SQL inner join which 
        merges dataframes based on similar values between both datasets
        
        3. left: The left method acts just like the SQL left join which merges 
        dataframes based on similar values from the left dataset
        
        4. right: The right method acts just like the SQL right join which
        dataframes based on similar values from the right dataset
